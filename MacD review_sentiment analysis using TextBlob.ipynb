```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob
```

```python
# Load the dataset with specified encoding
df = pd.read_csv("McDonald_s_Reviews.csv", encoding="latin-1")
```

```python
df.head()
```

```python
df['sentiment'] = df['review'].apply(lambda x :TextBlob(x).sentiment.polarity)
```

```python
df['sentiment']
```

```python
avg_sent = df.groupby('rating')['sentiment'].mean()
```

```python
avg_sent
```

```python
top_ten_sent = df.nlargest(10, 'sentiment')
```

```python
top_neg_sent = df.nsmallest(10, 'sentiment')
```

```python
top_ten_sent
```

```python
top_neg_sent
```

```python
print("Average sentiment by rating:")
print(avg_sent)
```

```python
print("\nTop 10 positive reviews:")
print(top_ten_sent[['store_name', 'review', 'rating']])
```

```python
print("\nTop 10 negative reviews:")
print(top_neg_sent[['store_name', 'review', 'rating']])
```

```python
color_palette = sns.color_palette("coolwarm", len(avg_sent))

plt.figure(figsize=(10, 6))
sns.barplot(x=avg_sent.index, y=avg_sent.values, palette=color_palette)
plt.xlabel('Rating')
plt.ylabel('Average Sentiment')
plt.title('Average Sentiment by Rating')
plt.show()
```
